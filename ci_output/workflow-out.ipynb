# Example workflow to calculate the travel time from one place to another in a specific vehicle

```python
from pyiron_workflow import Workflow
```

```python
from nodes.distance import get_distance
from nodes.vehicle import get_vehicle
from nodes.speed import get_speed
from nodes.time import get_time
```

```python
start = "Berlin"
end = "Paris"
vehicle = "train"
```

```python
wf = Workflow("calculate_travel_time")

wf.vehicle = get_vehicle(vehicle)
wf.distance = get_distance(start, end, wf.vehicle)
wf.speed = get_speed(wf.vehicle)
wf.time = get_time(wf.distance, wf.speed)
```

```python
wf.draw()
```

```python
wf()
```
