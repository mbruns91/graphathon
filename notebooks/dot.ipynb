```python
workflow = """
graph travel_time{
    input -> get_distance [label="start: str"];
    input -> get_distance [label="end: str"];
    input -> get_vehicle [label="vehicle: str"];
    get_vehicle -> get_speed [label="vehicle: str"];
    get_speed -> get_time [label="speed: float"];
    get_vehicle -> get_distance [label="vehicle: str"];
    get_distance -> get_time [label="distance: float"];
    get_time -> output [label="time: float"];
}
"""
```

```python
from graphathon.dot import EmptyGraph
```

```python
eg = EmptyGraph(workflow)
```

```python
eg.draw()
```

```python
eg.export()
```

```python
from pyiron_workflow import Workflow


wf = Workflow("travel_time")

vehicle = "Car"
start = "Berlin"
end = "Paris"

wf.vehicle = nodes.get_vehicle(vehicle)
wf.speed = nodes.get_speed(wf.vehicle)
wf.distance = nodes.get_distance(start, end, wf.vehicle)
wf.time = nodes.get_time(wf.speed, wf.distance)
```
